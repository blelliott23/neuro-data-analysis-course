```python
# Import the Pandas library
import pandas as pd 
```

### Let's load some data obtained from the NeuroElectro project at https://neuroelectro.org/api/docs

```python
file_name = "data/article_ephys_metadata_curated.csv"
pd.read_csv(file_name,delimiter='\t') # Load and display the data as a dataframe
```

```python
df = pd.read_csv(file_name,delimiter='\t',index_col='Index') # load into a dataframe using the 'Index' column as an index, 
                                                             # and store as a variable called 'df'
df = df.sort_index() # Sort this dataframe
df.head() # Show only the first 5 (or however many you want) rows
```

### That looks busy!  Can we simplify the output?

```python
columns = list(df)
sorted(columns)
```

```python
my_columns = ['BrainRegion','NeuronName','Species','rmp']
my_df = df[my_columns]
my_df.head(15)
```

### What is rmp, anyway?

![missing figure](images/membranepotential.gif)

### Now let's aggregate the data by neuron name and species

```python
means = my_df.groupby(['NeuronName','Species']).mean()
means.head(15)
```

### Let's make the column name more meaningful and also round to the nearest tenth of a mV

```python
means['rmp'] = means['rmp'].round(2)
means = means.rename(columns={'rmp':'Resting Membrane Potential'})
means.head(15)
```

### Compute the maximum value in the dataset

```python
max_vm = means['Resting Membrane Potential'].max()
print("The maximum resting membrane potential is %.1f mV" % max_vm)
```

### Let's get rid of the missing data

```python
means_nonans = means.dropna()
means_nonans.head(15)
```

### How about a different view?  

```python
means.unstack()
```

### Now the other way

```python
means.unstack().transpose()
```

### Let's make it more readable

```python
means.unstack().fillna('')
```

### You can chain all of this together into one line!

```python
pd.read_csv(file_name,delimiter='\t',index_col='Index').sort_index()[sorted(my_columns)]\
  .groupby(['NeuronName','Species']).mean().round(2).rename(columns={'rmp':'Resting Membrane Potential'})\
  .dropna().unstack().transpose().fillna('')
```
