# The mystery of the ans4 dataset

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)
```

```python
# Load the data
df = pd.read_csv('https://raw.githubusercontent.com/rgerkin/PythonDataScienceHandbook/master/notebooks/data/ans4.csv')

# Organize by group
df = df.set_index(['dataset','id'])

# Look at the first few rows
df.head()
```

```python
group = 'I'
df.loc[group].plot.scatter(x='x',y='y',c='k');
```

```python
group = 'II'
df.loc[group].plot.scatter(x='x',y='y',c='k');
```

```python
groups = ['I','II','III','IV'] # The list of group names
fig, axes = plt.subplots(2,2,figsize=(10,10)) # Make a canvas for the plots
for group in groups: # Iterate over each group
    i = groups.index(group) # Obtain the index (0, 1, 2, 3) of the current group
    ax = axes.flat[i]
    sns.scatterplot(x='x',y='y',data=df.loc[group], ax=ax, s=100) # Make a scatter plot
    ax.set_title('Group %s' % group)
plt.tight_layout() # Fix the margins so each plot has some breathing room
```

## Every group ($x$ and $y$) has: 
* ### Mean of $x = 9$
* ### Variance of $x = 11$
* ### Mean of $y = 7.5$
* ### Variance of $y = 4.125$

## Correlation between x and y for every group:
* ### $R = 0.816$

## Linear regression line: 
* ### $y = 0.5x + 3$
